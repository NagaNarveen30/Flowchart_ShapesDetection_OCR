```python
#model.py

import os
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
import time


def ocr_model():
    image_path = '/content/drive/MyDrive/dataset_flowchart/test/1.png'

    original_image=cv2.imread(image_path)
    #cv2.imshow(original_image)
    #resized_image=cv2.resize(original_image,(1920,1080))
    gray_image = cv2.cvtColor(original_image,cv2.COLOR_BGR2GRAY)

    # display images
    # create figure
    fig = plt.figure(figsize=(10, 7))

    # setting values to rows and column variables
    rows = 1
    columns = 2

    # Adds a subplot at the 1st position
    fig.add_subplot(rows, columns, 1)

    # showing image
    plt.imshow(original_image)
    plt.axis('off')
    plt.title("Original Image")

    # Adds a subplot at the 2nd position
    fig.add_subplot(rows, columns, 2)

    # showing image
    plt.imshow(gray_image)
    plt.axis('off')
    plt.title("Processed Image")

    plt.show()



    reader = easyocr.Reader(['en'])

    start_time = time.time()
    result = reader.readtext(gray_image,paragraph="False")
    end_time = time.time()
    elapsed_time = end_time - start_time

    font = cv2.FONT_HERSHEY_SIMPLEX
    text = []
    top_left = []
    bottom_right = []

    for i in range(len(result)):
        text.append(result[i][1])
        top_left.append(result[i][0][0])
        bottom_right.append(result[i][0][2])

    for i in range(len(result)):
        img = cv2.rectangle(original_image,top_left[i],bottom_right[i],(0,255,0),3)
        img = cv2.putText(original_image,text[i],bottom_right[i],font,0.5,(0,255,0),2,cv2.LINE_AA)

    width_screen,height_screen = 1920 , 1080
    width_img,height_img=original_image.shape[1],original_image.shape[0]
    x_scale=width_screen/width_img
    y_scale=height_screen/height_img

    # midpoint_x = 0
    # midpoint_y = 0

    # target = "Defect View"

    # for i in range (len(text)):
    #     if text[i] == target:
    #         midpoint_x=(top_left[i][0]+bottom_right[i][0])/2
    #         midpoint_y=(top_left[i][1]+bottom_right[i][1])/2
    #         break

    # final_x_coordinate = int(midpoint_x * x_scale)
    # final_y_coordinate = int(midpoint_y * y_scale)

    midpoint_x = [((top_left[i][0] + bottom_right[i][0]) / 2) * x_scale for i in range(len(text))]
    midpoint_y = [((top_left[i][1] + bottom_right[i][1]) / 2) * y_scale for i in range(len(text))]



    d = {'text': text, 'midpoint_x': midpoint_x, 'midpoint_y': midpoint_y}
    time_taken = {'elapsed_time' : elapsed_time}

    return d, time_taken









```
